**4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.**

```python
sns.displot(data=honey,x='priceperlb',kde=True,color='brown')
plt.title('Distribtion based on price of honey per pound')
plt.xlabel('price per pound')
plt.show()
```

**Observations**

1.  when the price per pound of honey is around 1.5$ , the distribution is max

2.  It shows that price per_lb in the range of 1 to 1.5 has the maximum count
 
3.  Hence the count is in decreasing order after crossing the 1.5$ mark

4.  The distribution is very low when the price per pound is around 4$

5. So it is evident that higher rate honey production is reducing and people prefer the budget friendly ones

**5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and
write your findings.**

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=honey,x='numcol',y='prodvalue',size='numcol',color='k')
plt.xlabel('Number of honey producing colonies',fontsize=13)
plt.ylabel('Value of production ',fontsize=13)
plt.title('Relationship between numcol and prod-value',fontsize=13)
plt.show()
```

**Observations**

1. we can see that the number of colonies  and value of production has a linear relationship
2. when the number of honey producing colony increases , the production value also increases and viceversa
3. The production value is max when the number of colonies are 500000
4. The production value is  zero , when the number of colonies are 0

**6. Plot and understand the relationship between categorical variable 'year' and a numerical variable
'prodvalue' through boxplot, and write your findings.**

```python
plt.figure(figsize=(18,6))
sns.boxplot(data=honey,y= 'prodvalue',x='year',)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Production value',fontsize=15)
plt.grid()
plt.show()
```

**Observations**
1. Each year has outliers in their production value
2. The production value distribution is  right skewed for almost all the years from 1998 to 2012
3. the Maximum production value comes in the year 2010 as an outlier followed by 2012
4. The max production value comes in the year 2002 and 2008 without considering the outliers 

**7. Visualize and understand the relationship between the multiple pairs of variables throughout different years
using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')**

```python
plt.figure(figsize=(15,10))
sns.pairplot(data=honey , vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue'],hue='year')
plt.show()
```

**Inferences**

2. The total production and number of colonies has  positive linear relationship
3. The  total production and production value has linear relationship.
4. Yeild per colony almost has a normal distribution throughout the  years 
5. Total production and number of colonies have a right skewed distribution 
6. Distribution of production value has a sharp peak , which indicates very less variance over the years 

**8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol',
'total prod', 'stocks', 'price per lb', 'prodvalue')**

```python
#displaying correlation between variables using heatmap
plt.figure(figsize=(8,5))
honey1 = honey[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']]
sns.heatmap(honey1.corr(),annot=True,annot_kws={'size':10},cmap='Greens')
```

**Inferences**

1. The number of colonies and total production has strong positive correlation of 0.95
2. The number of colonies and production value has a strong positive correlation of 0.91
3. the number of colonies and stocks has a strong positive correlation of 0.83
4. total production and stocks has a strong positive correlation of 0.88
5. Total production and production value has strong positive correlation of 0.91
6. stocks and production value has strong  positive correlation of 0.73
7. price per pound and production value has very weak correlation of -0.09 , which is almost zero
   it indicates that there is no linear relationship between these two variables 
8. All other variable associations are weak for analysis
