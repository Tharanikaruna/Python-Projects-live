**7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'
and so on.....**

```python
#creating a dictionary with month numbers as keys and month names as values
month_names= pd.Series({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})
```

```python
pd.DataFrame(month_names)
```

```python
#assigning the month names to their respective month numbers

df['published_month'] = df['published_month'].apply(lambda m:month_names[m])
# using lambda function we're assigning the name of months to their respective numbers 
```

```python
df[['published_month']] # the month numbers are converted to month names 
```

**8. Find the number of videos published each month and arrange the months in a decreasing order based

**13. Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.**

```python
# using scatter plot to show the relationship between dislikes and view_count variables

df.plot(kind = 'scatter',x='view_count',y='dislikes',figsize=(20,5),s=40,fontsize=15) 

#figsize - for adjusting width and height of the plot
#size (s) - to increase the size of scatter plots
#fontsize - to adjust the size of fonts on x and y axis 
```

```python
# Getting the view counts and its respective dislikes in tabular format 

df[['dislikes','view_count']].sort_values(by='view_count',ascending=False)
```

**Inferences**

From the above plot and table it is evident that the number of dislikes is directly influenced by the number of views

For videos with more dislikes , the number of views is more

For videos with less dislikes , the number of views is less

In very few cases there is an exception as we can see in the scatter plot, where the dislikes are more even with very less view counts 

But in general we can conclude that number of dislikes is directly proportional to the number of views
